```python
# =========================================================
# TASK 7 : DATA VISUALIZATION WITH TABLEAU
# COMPLETE PYTHON PROJECT CODE
# =========================================================

# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ---------------------------------------------------------
# STEP 1 : CREATE SAMPLE DATASET
# ---------------------------------------------------------

# Random seed for same output every time
np.random.seed(42)

# Number of records
records = 300

# Create dataset
data = {
    "Order_ID": range(1001, 1001 + records),

    "Region": np.random.choice(
        ["North", "South", "East", "West"],
        records
    ),

    "Category": np.random.choice(
        ["Electronics", "Furniture", "Clothing"],
        records
    ),

    "Sales": np.random.randint(
        500, 10000, records
    ),

    "Profit": np.random.randint(
        100, 3000, records
    ),

    "Quantity": np.random.randint(
        1, 15, records
    ),

    "Month": np.random.choice(
        ["January", "February", "March",
         "April", "May", "June"],
        records
    )
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Display first 5 rows
print("\nSample Dataset:\n")
print(df.head())

# Save dataset as CSV
df.to_csv("tableau_sales_data.csv", index=False)

print("\nCSV file 'tableau_sales_data.csv' created successfully!")

# ---------------------------------------------------------
# STEP 2 : BASIC DATA ANALYSIS
# ---------------------------------------------------------

print("\nDataset Information:\n")
print(df.info())

print("\nStatistical Summary:\n")
print(df.describe())

# ---------------------------------------------------------
# STEP 3 : BAR CHART - SALES BY REGION
# ---------------------------------------------------------

region_sales = df.groupby("Region")["Sales"].sum()

plt.figure(figsize=(8,5))
region_sales.plot(kind='bar')

plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")

plt.tight_layout()
plt.show()

# ---------------------------------------------------------
# STEP 4 : LINE CHART - MONTHLY SALES TREND
# ---------------------------------------------------------

month_sales = df.groupby("Month")["Sales"].sum()

# Arrange months properly
month_order = ["January", "February", "March",
               "April", "May", "June"]

month_sales = month_sales.reindex(month_order)

plt.figure(figsize=(8,5))

plt.plot(month_sales.index,
         month_sales.values,
         marker='o')

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")

plt.grid(True)

plt.tight_layout()
plt.show()

# ---------------------------------------------------------
# STEP 5 : PIE CHART - PROFIT BY CATEGORY
# ---------------------------------------------------------

category_profit = df.groupby("Category")["Profit"].sum()

plt.figure(figsize=(7,7))

plt.pie(
    category_profit,
    labels=category_profit.index,
    autopct='%1.1f%%'
)

plt.title("Profit Distribution by Category")

plt.tight_layout()
plt.show()

# ---------------------------------------------------------
# STEP 6 : DASHBOARD INSIGHTS
# ---------------------------------------------------------

print("\n================ DASHBOARD INSIGHTS ================\n")

print("1. Region-wise sales comparison completed.")
print("2. Monthly sales trend visualization completed.")
print("3. Profit distribution across categories completed.")
print("4. Dataset ready for Tableau dashboard creation.")

print("\n====================================================")

# ---------------------------------------------------------
# STEP 7 : HOW TO IMPORT INTO TABLEAU
# ---------------------------------------------------------

print("\nSTEPS TO USE IN TABLEAU:")
print("1. Open Tableau")
print("2. Click Connect -> Text File")
print("3. Select 'tableau_sales_data.csv'")
print("4. Create Bar Chart, Line Chart, and Pie Chart")
print("5. Assemble them into a Dashboard")
print("6. Save the Tableau Workbook")

# =========================================================
# END OF PROJECT
# =========================================================
```
