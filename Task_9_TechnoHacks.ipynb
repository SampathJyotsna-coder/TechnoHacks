```python
# ============================================
# TASK 9: INTRODUCTION TO DATA ANALYSIS CONCEPTS
# ============================================

# Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# --------------------------------------------
# STEP 1: CREATE SAMPLE DATASET
# --------------------------------------------

data = {
    "Student_Name": ["Asha", "Ravi", "Priya", "Kiran", "Meena",
                     "Arjun", "Sneha", "Rahul", "Divya", "Vikram"],

    "Age": [20, 21, 19, 22, 20, 21, 20, 23, 19, 22],

    "Marks": [85, 78, 92, 74, 88, 90, 84, 76, 95, 89],

    "Study_Hours": [4, 3, 6, 2, 5, 6, 4, 3, 7, 5],

    "Department": ["CSE", "ECE", "CSE", "EEE", "CSE",
                   "ECE", "IT", "EEE", "IT", "CSE"]
}

# Convert dictionary into DataFrame
df = pd.DataFrame(data)

# Display Dataset
print("\n===== STUDENT DATASET =====")
print(df)
print("\nDataset Shape:")
print(df.shape)
print("\nMissing Values:")
print(df.isnull().sum())
plt.figure(figsize=(6,4))

sns.heatmap(df[["Age", "Marks", "Study_Hours"]].corr(),
            annot=True,
            cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
print("\nFinal Insight:")
print("Students who study more tend to score higher marks.")
print("Data analysis helps identify trends and make decisions.")

# --------------------------------------------
# STEP 2: UNDERSTANDING DATA TYPES
# --------------------------------------------

print("\n===== DATA TYPES =====")
print(df.dtypes)

# Explanation
print("\nExplanation:")
print("int64  -> Numerical Integer Data")
print("object -> Text/Categorical Data")

# --------------------------------------------
# STEP 3: DESCRIPTIVE STATISTICS
# --------------------------------------------

print("\n===== DESCRIPTIVE STATISTICS =====")
print(df.describe())

# --------------------------------------------
# STEP 4: MEAN, MEDIAN, MODE
# --------------------------------------------

print("\n===== MEAN =====")
print(df["Marks"].mean())

print("\n===== MEDIAN =====")
print(df["Marks"].median())

print("\n===== MODE =====")
print(df["Marks"].mode())

# --------------------------------------------
# STEP 5: FIND MAXIMUM AND MINIMUM
# --------------------------------------------

print("\n===== HIGHEST MARKS =====")
print(df["Marks"].max())

print("\n===== LOWEST MARKS =====")
print(df["Marks"].min())

# --------------------------------------------
# STEP 6: DATA VISUALIZATION
# --------------------------------------------

# Set style
sns.set(style="whitegrid")

# BAR CHART
plt.figure(figsize=(8,5))
sns.barplot(x="Student_Name", y="Marks", data=df)

plt.title("Student Marks Analysis")
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.xticks(rotation=45)

plt.show()

# HISTOGRAM
plt.figure(figsize=(7,5))
plt.hist(df["Marks"], bins=5)

plt.title("Distribution of Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")

plt.show()

# SCATTER PLOT
plt.figure(figsize=(7,5))
sns.scatterplot(x="Study_Hours", y="Marks", data=df)

plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")

plt.show()

# --------------------------------------------
# STEP 7: REAL-WORLD INSIGHTS
# --------------------------------------------

print("\n===== REAL-WORLD ANALYSIS =====")

average_marks = df["Marks"].mean()

if average_marks > 80:
    print("Students are performing well academically.")
else:
    print("Students need improvement in academics.")

# Correlation
correlation = df["Study_Hours"].corr(df["Marks"])

print("\nCorrelation between Study Hours and Marks:")
print(correlation)

if correlation > 0:
    print("As study hours increase, marks also increase.")
else:
    print("Study hours and marks are negatively related.")

# --------------------------------------------
# STEP 8: CONCLUSION
# --------------------------------------------

print("\n===== CONCLUSION =====")
print("This project demonstrated:")
print("1. Data Types")
print("2. Descriptive Statistics")
print("3. Mean, Median, Mode")
print("4. Data Visualization")
print("5. Real-world Insights using Data Analysis")
```
