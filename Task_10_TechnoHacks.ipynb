```python
# =========================================================
# TASK 10 : EXPLORING DATA INSIGHTS
# =========================================================

# Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ---------------------------------------------------------
# STEP 1 : CREATE SAMPLE DATASET
# ---------------------------------------------------------

# Set random seed
np.random.seed(50)

# Number of records
records = 500

# Create sample sales dataset
data = {
    "Order_ID": range(1001, 1001 + records),

    "Region": np.random.choice(
        ["North", "South", "East", "West"],
        records
    ),

    "Category": np.random.choice(
        ["Electronics", "Furniture", "Clothing"],
        records
    ),

    "Sales": np.random.randint(
        1000, 15000, records
    ),

    "Profit": np.random.randint(
        100, 5000, records
    ),

    "Quantity": np.random.randint(
        1, 20, records
    ),

    "Month": np.random.choice(
        ["January", "February", "March",
         "April", "May", "June"],
        records
    )
}

# Convert into DataFrame
df = pd.DataFrame(data)

# Save dataset as CSV
df.to_csv("sales_insights_dataset.csv", index=False)

print("\nDataset Created Successfully!\n")

# Display first rows
print(df.head())

# ---------------------------------------------------------
# STEP 2 : BASIC DATA ANALYSIS
# ---------------------------------------------------------

print("\n================ DATASET INFORMATION ================\n")
print(df.info())

print("\n================ STATISTICAL SUMMARY ================\n")
print(df.describe())

# ---------------------------------------------------------
# STEP 3 : FIND TOTAL SALES BY REGION
# ---------------------------------------------------------

region_sales = df.groupby("Region")["Sales"].sum()

print("\n================ SALES BY REGION ================\n")
print(region_sales)

# Bar Chart
plt.figure(figsize=(8,5))

region_sales.plot(kind='bar')

plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")

plt.tight_layout()
plt.show()

# ---------------------------------------------------------
# STEP 4 : MONTHLY SALES TREND
# ---------------------------------------------------------

month_sales = df.groupby("Month")["Sales"].sum()

month_order = [
    "January", "February", "March",
    "April", "May", "June"
]

month_sales = month_sales.reindex(month_order)

print("\n================ MONTHLY SALES TREND ================\n")
print(month_sales)

# Line Chart
plt.figure(figsize=(8,5))

plt.plot(
    month_sales.index,
    month_sales.values,
    marker='o'
)

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.grid(True)

plt.tight_layout()
plt.show()

# ---------------------------------------------------------
# STEP 5 : PROFIT DISTRIBUTION BY CATEGORY
# ---------------------------------------------------------

category_profit = df.groupby("Category")["Profit"].sum()

print("\n================ PROFIT BY CATEGORY ================\n")
print(category_profit)

# Pie Chart
plt.figure(figsize=(7,7))

plt.pie(
    category_profit,
    labels=category_profit.index,
    autopct='%1.1f%%'
)

plt.title("Profit Distribution by Category")

plt.tight_layout()
plt.show()

# ---------------------------------------------------------
# STEP 6 : IDENTIFY KEY INSIGHTS
# ---------------------------------------------------------

highest_sales_region = region_sales.idxmax()
highest_sales_value = region_sales.max()

highest_profit_category = category_profit.idxmax()

best_month = month_sales.idxmax()

print("\n================ KEY INSIGHTS ================\n")

print(f"1. Highest Sales Region : {highest_sales_region}")
print(f"2. Total Sales in Highest Region : {highest_sales_value}")

print(f"\n3. Most Profitable Category : {highest_profit_category}")

print(f"\n4. Best Performing Month : {best_month}")

print("\n5. Sales trend shows business growth patterns.")

print("\n====================================================")

# ---------------------------------------------------------
# STEP 7 : SAVE SUMMARY REPORT
# ---------------------------------------------------------

report = f"""
=================================================
        SALES DATA ANALYSIS REPORT
=================================================

1. Highest Sales Region:
   {highest_sales_region}

2. Total Sales in Highest Region:
   {highest_sales_value}

3. Most Profitable Category:
   {highest_profit_category}

4. Best Performing Month:
   {best_month}

5. Key Findings:
   - Regional sales performance analyzed
   - Monthly trends identified
   - Profit contribution by category visualized
   - Business insights generated successfully

=================================================
"""

# Save report to text file
with open("sales_analysis_report.txt", "w") as file:
    file.write(report)

print("\nReport Generated Successfully!")
print("\nFile Saved : sales_analysis_report.txt")

# =========================================================
# END OF PROJECT
# =========================================================
```
